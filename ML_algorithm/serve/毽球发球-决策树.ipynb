# 毽球发球-决策树

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeClassifier
```

```python
fq = pd.read_excel('fq_DataSet.xlsx')
```

# 确定特征值和目标值

```python
data = fq.iloc[:,:44]

target = fq['L']

display(data,target)
```

# 将数据划分成训练集和验证集

```python
#打乱顺序，按照比例进行划分
from sklearn.model_selection import train_test_split
```

```python
#test_size = 0.2. train_size = 0.8
#训练：测试 = 4:1
data_train,data_test,target_train,target_test = train_test_split(data,target,test_size = 0.2,random_state= 90)
```

```python
estimator = DecisionTreeClassifier(criterion = "entropy", max_depth = 2)
estimator.fit(data_train, target_train)
```

```python
estimator.score(data_test, target_test)
```

# 决策树可视化

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(estimator, out_file = "tree.dot")
```
