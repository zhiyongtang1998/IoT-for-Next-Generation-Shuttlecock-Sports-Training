```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
#算法knn
#蹴回旋球  专业组 576次 标签 1
#        业余组 612次 标签 0
```

```python
fq = pd.read_excel('ALL_DataSet.xlsx')
```

```python
#前44列 为特征 data L为target
```

```python
data = fq.iloc[:,:44]

target = fq['L']

display(data,target)
```

```python
#打乱顺序，按照比例进行划分
from sklearn.model_selection import train_test_split
```

```python
#test_size = 0.2. train_size = 0.8
#训练：测试 = 4:1
data_train,data_test,target_train,target_test = train_test_split(data,target,test_size = 0.3,random_state= 90)
```

```python
accuracy = 0
for i in range(50):
    #算法，训练
    knn = KNeighborsClassifier(n_neighbors=9)
    knn.fit(data_train,target_train)

    #使用算法进行预测
    #
    target_ = knn.predict(data_test)

    #准确率
    accuracy += (target_ == target_test).mean()/50
print('---------多次训练平均准确率：%0.3f'%(accuracy))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(target_test,target_)
```

```python
from sklearn.metrics import classification_report

print(classification_report(target_test,target_,target_names = ['脚内侧踢球','正脚背发球']))
```
