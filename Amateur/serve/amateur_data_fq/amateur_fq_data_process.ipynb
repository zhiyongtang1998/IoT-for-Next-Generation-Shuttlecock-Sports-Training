```python
import pandas as pd
data_fq_A = pd.read_excel('amateur_data_fq_A.xlsx',index_col='index')
```

```python
data_fq_A.reset_index(inplace=True,drop=True)
```

```python
data_fq_A
```

```python
data_fq_B_columns=[]
for i in range(50):
    for column in data_fq_A.columns:
        data_fq_B_columns.append(column+'_'+str(i))
print(data_fq_B_columns)
```

```python
data_fq_B=pd.DataFrame(columns=data_fq_B_columns)
```

```python
data_fq_B
```

```python
for i in range(int(len(data_fq_A)/50)):
    data_columns = []
    for j in range(50):
        data_columns.extend(list(data_fq_A.loc[i*50+j]))
    data_fq_B.loc[i] = data_columns
    data_columns.clear()
```

```python
data_fq_B
```

```python
data_fq_B_T = pd.DataFrame(data_fq_B.values.T, index=data_fq_B.columns, columns=data_fq_B.index)#转置
```

```python
data_fq_B_T
```

```python
data_fq_B_T.to_excel('amateur_data_fq_B_T.xlsx')
```
